```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt  
```

```python
# 读入数据 
data = genfromtxt(r"longley.csv",delimiter=',')
print(data)
```

```python
# 切分数据
x_data = data[1:,2:]
y_data = data[1:,1,np.newaxis]
print(x_data)
print(y_data)
```

```python
print(np.mat(x_data).shape)
print(np.mat(y_data).shape)
# 给样本添加偏置项 
X_data = np.concatenate((np.ones((16,1)),x_data),axis=1) #生成一个全是1的numpyarray16行1列然后和x_data合并
print(X_data.shape)
```

```python
print(X_data[:3])
```

```python
# 岭回归标准方程法求解回归参数
def weights(xArr, yArr, lam=0.2):#0.2为岭系数的值，可以自己考虑什么数值比较好
    xMat = np.mat(xArr)
    yMat = np.mat(yArr)
    xTx = xMat.T*xMat # 矩阵乘法
    rxTx = xTx + np.eye(xMat.shape[1])*lam #1这里指(16,7)的下标第一列，得到数字7 即生成了一个7行7列的单位矩阵
    # 计算矩阵的值,如果值为0，说明该矩阵没有逆矩阵
    if np.linalg.det(rxTx) == 0.0:
        print("This matrix cannot do inverse")
        return
    # xTx.I为xTx的逆矩阵
    ws = rxTx.I*xMat.T*yMat
    return ws
```

```python
ws = weights(X_data,y_data)
print(ws)
```

```python
# 计算预测值
np.mat(X_data)*np.mat(ws)
```
