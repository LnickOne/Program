```python
import numpy as np
from numpy import genfromtxt
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# 读入数据 
data = genfromtxt(r"longley.csv",delimiter=',')
print(data)
```

```python
# 切分数据
x_data = data[1:,2:]
y_data = data[1:,1]
print(x_data)
print(y_data)
```

```python
# 创建模型
# 生成50个值
alphas_to_test = np.linspace(0.001, 1) #岭回归里面入的值,生成50个候选值，然后看哪个更合适
# 创建模型，保存误差值
model = linear_model.RidgeCV(alphas=alphas_to_test, store_cv_values=True) #CV代表交叉验证，用交叉验证法测试这50个值 哪个比较好
model.fit(x_data, y_data)

# 岭系数
print(model.alpha_)
# loss值
print(model.cv_values_.shape)#16代表16个样本，交叉验证的时候会取15个样本做训练集，1个样本做测试集，会得到16个loss值,50代表岭系数
```

```python
# 画图
# 岭系数跟loss值的关系
plt.plot(alphas_to_test, model.cv_values_.mean(axis=0))#axis=0或者1代表求平均值的方向，等于0代表对16个loss值求平均，就是交叉验证法最后一步的样本/1
# 选取的岭系数值的位置
plt.plot(model.alpha_, min(model.cv_values_.mean(axis=0)),'ro')
plt.show()
```

```python
model.predict(x_data[2,np.newaxis])
```
