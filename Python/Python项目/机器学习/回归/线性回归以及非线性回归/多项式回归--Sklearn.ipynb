```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# 载入数据
data = np.genfromtxt("job.csv", delimiter=",")
x_data = data[1:,1] #下标第一行到最后一行加上第一列   逗号,之前全是行的取值  逗号,之后才是列取值
y_data = data[1:,2]#下标第一行到最后一行加上第二列
plt.scatter(x_data,y_data)
plt.show()
```

```python
#x_data
```

```python
x_data = x_data[:,np.newaxis] #np.newaxis是加维度的意思
y_data = y_data[:,np.newaxis]
```

```python
#x_data
```

```python
#y_data
```

```python
# 创建并拟合模型
model = LinearRegression()#这是一元回归模型
model.fit(x_data, y_data) #创造一条直线拟合数据,传入的是二维数据，即每一行都是数据
```

```python
# 画图
plt.plot(x_data, y_data, 'b.')#样式为点图，蓝色
plt.plot(x_data, model.predict(x_data), 'r')#预测值为蓝色,将x_data传入即为y的值,为红色
plt.show()
```

```python
# 定义多项式回归,degree的值可以调节多项式的特征
poly_reg  = PolynomialFeatures(degree=5) 
# 特征处理
x_poly = poly_reg.fit_transform(x_data) #调用fit_transform方法,将x_data处理成多项式
# 定义回归模型
lin_reg = LinearRegression()
# 训练模型
lin_reg.fit(x_poly, y_data) #用特征处理后的数据训练
```

```python
#x_poly
```

```python
x_test = np.linspace(1,10,50)
x_test = x_test[:,np.newaxis]
```

```python
# 画图
plt.plot(x_data, y_data, 'b.')#原始数据
x_test = np.linspace(1,10,100)
x_test = x_test[:,np.newaxis]
plt.plot(x_test, lin_reg.predict(poly_reg.fit_transform(x_test)), c='r')#拟合数据
#plt.plot(x_data, lin_reg.predict(x_poly), c='r')  #拟合数据,传入特征处理好的数据

plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
