```python
import numpy as np
from numpy import genfromtxt
from sklearn import linear_model
import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D 
```

```python
# 读入数据 
data = genfromtxt(r"Delivery.csv",delimiter=',')
print(data)
```

```python
# 切分数据
x_data = data[:,:-1]#所有的行和第一列做训练数据
y_data = data[:,-1]#所有的行和最后一列做标签（值）
print(x_data)
print(y_data)
```

```python
# 创建模型
model = linear_model.LinearRegression()#实例化线性回归模型，这个模型是二元线性回归模型
model.fit(x_data, y_data)#调用fit方法（函数）传入x_data和y_data，就把模型参数训练好了，传入两个特征就得到两个系数，三个特征就三个系数类推
```

```python
#用coef_和intercept_查看系数和截距
# 系数

print("coefficients:",model.coef_)

# 截距(又叫偏置项)
print("intercept:",model.intercept_)

# 测试
x_test = [[109,4]]
predict = model.predict(x_test)#测试一个数据传入predict
print("predict:",predict)
```

```python
ax = plt.figure().add_subplot(111, projection = '3d') 
ax.scatter(x_data[:,0], x_data[:,1], y_data, c = 'r', marker = 'o', s = 100) #点为红色三角形  ,画散点
x0 = x_data[:,0] #把两个数据切分一下
x1 = x_data[:,1]
# 生成网格矩阵
x0, x1 = np.meshgrid(x0, x1)
z = model.intercept_ + x0*model.coef_[0] + x1*model.coef_[1]  #和各自的系数相乘
# 画3D图
ax.plot_surface(x0, x1, z)
#设置坐标轴  
ax.set_xlabel('Miles')  
ax.set_ylabel('Num of Deliveries')  
ax.set_zlabel('Time')  
  
#显示图像  
plt.show()  
```
