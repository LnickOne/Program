```python
from sklearn.feature_extraction import DictVectorizer
from sklearn import tree
from sklearn import preprocessing
import csv
```

```python
# 读入数据
Dtree = open(r'buy_computer.csv', 'r')
reader = csv.reader(Dtree)

# 获取第一行数据
headers = reader.__next__()#这里是获得四个属性
print(headers) 

# 定义两个列表
featureList = [] #特征list
labelList = [] #标签list


for row in reader:
    labelList.append(row[-1])# 把所有标签存入特征list
    rowDict = {}#每读取一行建立一个数据字典
    for i in range(1, len(row)-1):#循环从1到5
        rowDict[headers[i]] = row[i] #将对应的特征属性键值匹配，例如headers[1]是age,row[1]是youth
    featureList.append(rowDict) # 把所有特征数据字典存入标签list

print(featureList)
```

```python
# 把数据转换成01表示
vec = DictVectorizer()
x_data = vec.fit_transform(featureList).toarray()#调用fit_transform方法
print("x_data: " + str(x_data)) #x_data如下图

# 打印第一行特征属性
print(vec.get_feature_names()) #10个列属性对应上面0 1，把每个特征的属性值全列出来，比如age=middle_aged', 'age=senior', 'age=youth

# 打印标签
print("labelList: " + str(labelList))

# 把标签转换成01表示,no是0,yes是1
lb = preprocessing.LabelBinarizer()
y_data = lb.fit_transform(labelList)
print("y_data: " + str(y_data))
```

```python
# 创建决策树模型
model = tree.DecisionTreeClassifier(criterion='entropy')#entropy代表的是信息熵，使用ID3或者C4.5算法
# 输入数据建立模型
model.fit(x_data, y_data)
```

```python
# 测试
x_test = x_data[0]
print("x_test: " + str(x_test))

predict = model.predict(x_test.reshape(1,-1))
a=x_test.reshape(1,-1)#二维化
print(a)
print("predict: " + str(predict))
```

```python
# 导出决策树
# pip install graphviz
# http://www.graphviz.org/
import graphviz

dot_data = tree.export_graphviz(model,
                                out_file=None,
                                feature_names=vec.get_feature_names(),
                                class_names=lb.classes_,
                                filled=True,
                                rounded=True,
                                special_characters=True)
# model,
#out_file = None,
#feature_names = vec.get_feature_names(), 这三个重要参数，model,特征名
graph = graphviz.Source(dot_data)
graph.render('computer')
```
