```python
import matplotlib.pyplot as plt
import numpy as np
import operator

# 已知分类的数据
x1 = np.array([3,2,1])
y1 = np.array([104,100,81])
x2 = np.array([101,99,98])
y2 = np.array([10,5,2])
scatter1 = plt.scatter(x1,y1,c='r')
scatter2 = plt.scatter(x2,y2,c='b')

# 未知数据
x = np.array([18])
y = np.array([90])
scatter3 = plt.scatter(x,y,c='k')


#画图例
plt.legend(handles=[scatter1,scatter2,scatter3],labels=['labelA','labelB','X'],loc='best')

plt.show()
```

```python
# 已知分类的数据
x_data = np.array([[3,104],
                   [2,100],
                   [1,81],
                   [101,10],
                   [99,5],
                   [81,2]])
y_data = np.array(['A','A','A','B','B','B'])
x_test = np.array([18,90])
```

```python
# 计算样本数量
x_data_size = x_data.shape[0]
x_data_size
```

```python
# 复制x_test,这样可好计算和这六个样本点的距离
np.tile(x_test, (x_data_size,1))
```

```python
# 计算x_test与每一个样本的差值
diffMat = np.tile(x_test, (x_data_size,1)) - x_data
diffMat
```

```python
# 计算差值的平方
sqDiffMat = diffMat**2
sqDiffMat
```

```python
# 求和
sqDistances = sqDiffMat.sum(axis=1)
sqDistances
```

```python
# 开方 ，欧式距离的公式
distances = sqDistances**0.5
distances
```

```python
# 从小到大排序，得到从小到大数据的索引排列
sortedDistances = distances.argsort()
sortedDistances
```

```python
classCount = {}
# 设置k
k = 5#找到它最近的五个点
for i in range(k):
    # 获取标签
    votelabel = y_data[sortedDistances[i]] #取到y_data里面第i个索引的标签
    # 统计标签数量
    classCount[votelabel] = classCount.get(votelabel,0) + 1 #如果标签这个键不存在 则它的值就赋值为0，如果有这个键了 就把它的值+1
```

```python
classCount
```

```python
# 根据operator.itemgetter(1)-第1个值对classCount排序，然后再取倒序
sortedClassCount = sorted(classCount.items(),key=operator.itemgetter(1), reverse=True) #对字典的第一个部分排序（即值，就是3 和2排序），True表示从大到小排序
sortedClassCount
```

```python
# 获取数量最多的标签
knnclass = sortedClassCount[0][0]
knnclass
```
